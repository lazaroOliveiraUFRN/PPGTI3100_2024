```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import KMeans,AgglomerativeClustering,DBSCAN,AffinityPropagation,OPTICS,MiniBatchKMeans,SpectralClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score

from sklearn_extra.cluster import KMedoids


import numpy as np


import scipy
import pandas as pd

from ipywidgets import IntProgress
from IPython.display import display
import time

from sklearn.preprocessing import MaxAbsScaler

from sklearn.decomposition import PCA,TruncatedSVD
from math import pi

import html
```

```python
lista_embed = []
for idx, palavra in enumerate(lista_produtos_embed):
    e={}
    for idx2, embed in enumerate(palavra['embedding']):
        e['componente'+str(idx2)] = embed

    lista_embed.append(e)
```

```python
df = pd.DataFrame(lista_embed)
df
```

```python
X=np.nan_to_num(df.values)
faixa=range(2,35,5)
cluster_inertia=[]
for k in faixa:
    model=KMeans(n_clusters=k,init='k-means++', max_iter=30, n_init=751, verbose=0)
    #model=MiniBatchKMeans(n_clusters = idx, batch_size=128)
    model.fit(X)
    Y=model.labels_
    value_silhouette_score=0
    value_silhouette_score=silhouette_score(X,Y)
    value_davies_bouldin_score = davies_bouldin_score(X, Y)
    print(k,';',model.inertia_,value_silhouette_score,value_davies_bouldin_score)
    e={'cluster':k,'inertia':model.inertia_,'silhouette_score':value_silhouette_score, 'davies_bouldin_score':value_davies_bouldin_score}
    cluster_inertia.append(e)
```

```python
from matplotlib.ticker import MultipleLocator

fig, ax = plt.subplots(figsize=(15, 10))
k = faixa
plt.plot(k, [x['silhouette_score'] for x in cluster_inertia], 'bx-')
plt.xlabel('Número de  clusters', fontsize = 10)
plt.ylabel('Elbow Method - silhouette score', fontsize = 10)
plt.title(f'Produtos - Kmeans')


ax.xaxis.set_major_locator(MultipleLocator(5))
#ax.yaxis.set_major_locator(MultipleLocator(0.05))

ax.spines["bottom"].set_color("black")
ax.spines["top"].set_color("black")
ax.spines["left"].set_color("black")
ax.spines["right"].set_color("black")

ax.minorticks_on()
plt.grid(visible=True,which='major',axis='both',zorder=3,linestyle='-')
plt.grid(visible=True,which='minor',axis='both',zorder=3,linestyle=':')

plt.show()
```
